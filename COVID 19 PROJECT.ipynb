```python
ab.columns=['State/UTs','Total_Cases','Active','Deaths','Active_Ratio','Death_Ratio']
```

```python
ab.groupby(['State/UTs'])['Active_Ratio'].agg(['max'])
```

```python
# plt.figure(dpi=200)
plt.figure(figsize=(10,4))
plt.plot(ab['State/UTs'],ab['Active_Ratio'],color='red',marker='o')
plt.xticks(rotation=90,fontsize=13.5)
plt.yticks(fontsize=18.5)
plt.xlabel('State', fontsize=15)

plt.ylabel('Active Ration', fontsize=15)
plt.title("Line Plot for Active Ratio",fontsize=15)
plt.show()

```

```python
plt.figure(figsize=(15,8));
plt.xticks(rotation=75)
sns.barplot(ab['State/UTs'],ab['Total_Cases']);
plt.title('total cases each state',fontsize=20);
# plt.xl/able('state',fontsize=15)
```

```python
plt.figure(figsize=(8,8))
labels = ab.index
plt.pie(x='Active', data=ab.head(), labels='State',startangle=90,autopct='%.1f%%')
plt.title("Active Cases in Top 5 states in India", fontsize = 24) 
plt.tight_layout() 
plt.show()

```
